```python
#Importimng the modules needed to open zipped files
from io import BytesIO
from zipfile import ZipFile
```

```python
#to make the network
import networkx as nx
```

```python
url = 'http://www-personal.umich.edu/~mejn/netdata/power.zip' 
```

```python
#following the path used in tutorials to open the files
with  urllib.urlopen(url) as stream:
  with ZipFile(BytesIO(stream.read())) as archive:
    archive.printdir() 
```

```python
  with  urllib.urlopen(url) as stream:
    with ZipFile(BytesIO(stream.read())) as archive: #extracted files as strings
      gml = archive.read('power.gml').decode()
      text = archive.read('power.txt').decode()
```

```python
#give me the nodes and edges
print(gml)
```

```python
#need to parse the gml to be able to draw it
gogo = nx.parse_gml(gml, label='id')
```

```python
#lets see what happens when we draw it
nx.draw_networkx(gogo)
```
