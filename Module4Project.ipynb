```python
thisurl = requests.get('https://phoenix.craigslist.org/d/garage-moving-sales/search/gms')
```

```python
#I want to parse the page so I will pass it through beautifulsoup
garage = BeautifulSoup(thisurl.text,'html.parser')
```

```python
garage
```

```python
#now i want to isolate the results.
listings = garage.find_all('li',class_='result-row')
```

```python
#a loop to get the links and titles of each post
listing_title_texts=[]
listing_links=[]

for listing in listings:
    listing_title=listing.find('a',class_='result-title hdrlnk')
    listing_title_text= listing_title.text
    listing_title_texts.append(listing_title_text)
    listing_link= listing_title['href']
    listing_links.append(listing_link)
print('\n')
```
